importing basic libraries

```python
import altair as alt
from sklearn.datasets import load_iris
import pandas as pd
```

Loading iris dataset

```python
data = load_iris()
```

```python
iris_df = pd.DataFrame(data=data['data'], columns=data['feature_names'])
```

```python
species_dict = {i:specie for i, specie in enumerate(data['target_names'])}
```

```python
iris_df['Species'] = data['target']
iris_df['Species'] = iris_df['Species'].apply(lambda x: species_dict[x])
```

```python
iris_df.head()
```

Testing Vega-Altair library

```python
chart = alt.Chart(iris_df)
```

```python
chart.mark_point()
```

```python
chart.mark_point().encode(
    x='Species',
    ).properties(
    width=300,
    height=100,
)
```

```python
chart.mark_point().encode(
    x='sepal length (cm)',
    y='sepal width (cm)'
    ).properties(
    width=400,
    height=400,
)
```

```python
chart.mark_point().encode(
    x='Species',
    y='average(sepal width (cm))'
)
```

```python
chart.mark_bar().encode(
    x='Species',
    y='average(sepal width (cm))'
)
```

```python
chart.mark_bar().encode(
    x='average(sepal width (cm))',
    y='Species',
)
```

```python
chart.mark_bar().encode(
    x='count(Species)',
    y='Species',
)
```

```python
chart.mark_bar().encode(
    alt.Y('Species', type='nominal'),
    alt.X('sepal width (cm)', type='quantitative', aggregate='average')
)
```

Changing Color

```python
chart.mark_bar(color='firebrick').encode(
    alt.Y('Species', type='nominal'),
    alt.X('sepal width (cm)', type='quantitative', aggregate='average').title('Avg. Sepal Width per Species')
).properties(
    width=400,
    height=200,
)
```
